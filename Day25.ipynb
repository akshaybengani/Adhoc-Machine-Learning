```python
# Starting Camera using 0 as the 1st ID
import cv2
import matplotlib.pyplot as plt
cap = cv2.VideoCapture(0)

while True:
    status,img = cap.read()
    grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # to show a image in Grey color
    cv2.imshow("window 1",img)
    cv2.imshow("window 2",grey) 
    if cv2.waitKey(10) == 13:   # This shit 10 is 10 milliseconds
        break

cv2.destroyAllWindows()
cap.release()
```

```python
# Adding a Rectangle in the program
import cv2
import matplotlib.pyplot as plt
cap = cv2.VideoCapture(0)

while True:
    status,img = cap.read()
    cv2.line(img,(0,0),(200,300),(0,255,0),3)
    cv2.rectangle(img,(50,50),(250,300),(255,0,0),3)
    cv2.imshow("window 1",img)
    if cv2.waitKey(10) == 13:   # This shit 10 is 10 milliseconds
        break

cv2.destroyAllWindows()
cap.release()
```

```python
# Adding a Circle in the program
import cv2
import matplotlib.pyplot as plt
cap = cv2.VideoCapture(0)

while True:
    status,img = cap.read()
    cv2.line(img,(0,0),(200,300),(0,255,0),3)
    cv2.rectangle(img,(50,50),(250,300),(255,0,0),3)
    cv2.circle(img,(200,300),100,(13,45,54),3)
    cv2.imshow("window 1",img)
    if cv2.waitKey(10) == 13:   # This shit 10 is 10 milliseconds
        break

cv2.destroyAllWindows()
cap.release()
```

```python
# Adding a Text in the program
import cv2
import matplotlib.pyplot as plt
cap = cv2.VideoCapture(0)

while True:
    status,img = cap.read()
    cv2.line(img,(0,0),(200,300),(0,255,0),3)
    cv2.rectangle(img,(50,50),(250,300),(255,0,0),3)
    cv2.circle(img,(200,300),100,(13,45,54),3)
    
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(img,'OpenCV',(10,300), font, 4,(255,255,255),10,cv2.LINE_4)
    
    cv2.imshow("window 1",img)
    if cv2.waitKey(10) == 13:   # This shit 10 is 10 milliseconds
        break

cv2.destroyAllWindows()
cap.release()
```

```python
# Writting a video
import cv2
import matplotlib.pyplot as plt
cap = cv2.VideoCapture(0)
plugin = cv2.VideoWriter_fourcc(*'XVID')
output = cv2.VideoWriter('video1.avi',plugin,80,(640,480))

while True:
    status,img = cap.read()
    cv2.imshow("window 1",img)
    output.write(img)
    #print(img.shape)
    if cv2.waitKey(10) == 13:   # This shit 10 is 10 milliseconds
        break

cv2.destroyAllWindows()
cap.release()
```
