```python
import pandas as pd
```

```python
df = pd.read_csv('http://13.234.66.67/summer19/datasets/social.csv')
```

```python
df
```

```python
df.info()
```

```python
filterdata = df.iloc[0:,2:].values
filterdata
```

```python
# Now we can apply training test
features = filterdata[0:,0:2]
```

```python
#Label Finding
label = filterdata[0:,2]
```

```python
# Importing training and Testing
from sklearn.model_selection import train_test_split
```

```python
train_features,test_features,train_label,test_label = train_test_split(features,label,test_size=0.2)
```

```python
# Now apply features Scalling
from sklearn.preprocessing import StandardScaler
```

```python
# This is called as feaure Scalling
sc = StandardScaler()
```

```python
# Apply scalling in training data only
sc.fit_transform(train_features)
```

```python
# testing data Scaling
sc.transform(test_features)
```

```python
# KNN
# ensemble means showing yourself repeatation of satadata
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Random forest call with number of decision Tree
Rclf = RandomForestClassifier(n_estimators=20)
```

```python
# Now Training
trained = Rclf.fit(train_features,train_label)
```

```python
output=trained.predict(test_features)
output
```

```python
test_label
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test_label,output)
```
