To import a specific function or section of a huge liberary we use "." operator to specify things

````python
import matplotlib.pyplot as plt
````

````python
x=[2,3]
y=[9,5]
plt.plot(x,y)
````

* The plot function draws a straight line between the points<br>
* In some cases we have to use plt.show() function to display the graph.<br>
* To add labels on axis we have functions called ```xlabel()``` and ```ylabel()```<br>
* To add grids in the graph use the function grid(color="COLORNAME")<br>
* To add labels to multiple lines or plots you can pass the label name with plot functon<br>
* To show labels we have a function called legend()
* To fix the size of xaxis and its differences we use the function xlim(StartingPoint,End point) vice versa yslim for yaxis.

````python
plt.xlabel('Time')
plt.ylabel('Speed')
plt.grid(color="Green")
plt.xlim(0,10)
plt.ylim(0,10)
plt.plot(x,y,label="water")
x1=[4,3,8,5]
y1=[2,9,7,7]
plt.plot(x1,y1,label="sand")
plt.legend()
````

````python
plt.bar(x,y,label='water')
plt.bar(x1,y1,label='sand')
plt.grid(color='green')
````
